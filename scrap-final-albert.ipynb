```python
#Для основной табл с wikicfp
def parse_main_table(link):
    date_start=''
    date_end=''
    city=''
    country=''
    ddln_sub=''
    ddln_not='' 
    ddln_fnl_ver=''
    fields=''
    main_info = next_scrap(link).findAll('table', {'class':'gglu'})[0].findAll('tr')
    for row in main_info:
        sib = row.findNext('th')
        if sib.text == 'When':
            dates = sib.findNext('td').text
            dates = dates.replace(',',' ')
            dates = ' '.join(dates.split())
            dates = dates.split()
            date_start = ' '.join(dates[0:3])
            date_start=parse(date_start)
            date_end=' '.join(dates[4:])
            date_end=parse(date_end)
        elif sib.text == 'Where':
            place = sib.findNext('td').text
            place = str(place)
            place = place.replace(',',' ')
            place = ' '.join(place.split())
            place = place.split(' ')
            if len(place)==2:
                city = place[0]
                country = place[1]
            elif len(place)==3:
                city = place[0]
                country = place[2]
        elif sib.text == 'Submission Deadline':
            ddln_sub = sib.findNext('td').text
            ddln_sub = str(ddln_sub)
            ddln_sub = ddln_sub.replace(',',' ')
            ddln_sub = ' '.join(ddln_sub.split())
            if 'TBD'in ddln_sub:
                ddln_sub='TBD'
            else:
                ddln_sub=parse(ddln_sub)
        elif sib.text == 'Notification Due':
            ddln_not = sib.findNext('td').text
            ddln_not = str(ddln_not)
            ddln_not = ddln_not.replace(',',' ')
            ddln_not = ' '.join(ddln_not.split())
            if 'TBD'in ddln_not:
                ddln_not='TBD'
            else:
                ddln_not=parse(ddln_not)
        elif sib.text == 'Final Version Due':
            ddln_fnl_ver = sib.findNext('td').text
            ddln_fnl_ver = str(ddln_fnl_ver)
            ddln_fnl_ver = ddln_fnl_ver.replace(',',' ')
            ddln_fnl_ver = ' '.join(ddln_fnl_ver.split())
            if 'TBD'in ddln_fnl_ver:
                ddln_fnl_ver='TBD'
            else:
                ddln_fnl_ver=parse(ddln_fnl_ver)

        else:
            pass
    fields= next_scrap(link).findAll('table', {'class':'gglu'})[1].text
    fields=fields.replace('\n','\xa0')
    fields=fields.split('\xa0')
    fields =list(filter(None, fields))
    fields=fields[1:]
    return(date_start,
date_end,
city,
country,
ddln_sub,
ddln_not, ddln_fnl_ver,fields)
```

```python
def get_titles(link):
    confTitle=''
    confFamily=''
    titles = next_scrap(link).findAll('center')[0].findAll('tr')[4].text
    titles=str(titles)
    titles=titles.replace('\n','\xa0')
    titles=titles.split('\xa0')
    titles =list(filter(None, titles))
    if len(titles)==2:
    
        confTitle=titles[0]
        confFamily=titles[1]
    else:
        confTitle=titles[0]
    return(confTitle,confFamily)
```

```python
def get_website_address(link):
    website=next_scrap(link).findAll('center')[0].findAll('tr')[5].find('a').get('href')
    website=str(website)
    return(website)
```

```python
file=open('list of all acm conferences.txt',mode='r')
data=[]
for line in file.readlines():
    a =line.split('\n')
    data.append(a)
names=[]
for name in data:
    for i in name:
        if '\t' in i:
            i=i.split('\t')
            names.append(i)
file.close() 
```

```python
mdf_names=[]
for a in names:
    number=a[0].find('logo')
    nm_acm=a[0][:number].split()
    last= nm_acm[-1]
    full=a[1].split(' ')
    for i,item in enumerate(full):
        if last in item:
            number=len(last)
            a, b=item[:number], item[number:]
            full[i]=[a,b]
            mdf_names.append(full)
            break
```

```python
def together(lists):
    new_list=[]
    for i in lists:
        if type(i)==list:
            for j in i:
                new_list.append(j)
        else:
            new_list.append(i)
    return(new_list)
```

```python
mdfi_names=[]
for i in mdf_names:
    mdfi_names.append(' '.join(together(i)))
```
